```python
import yfinance as yf
import pandas as pd

# GME hisse senedi verilerini indir
gme_data = yf.download('GME', start='2017-01-01', end='2023-01-01')

# İndeksi sıfırla
gme_data.reset_index(inplace=True)

# Yıllık bazda gelir benzeri bir veri türetme (örnek)
gme_data['Year'] = gme_data['Date'].dt.year
gme_revenue = gme_data.groupby('Year')['Volume'].sum().reset_index()

# Son 5 satırı göster
print(gme_revenue.tail())
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Tesla hisse senedi verilerini indir
tesla_data = yf.download('TSLA', start='2020-01-01', end='2023-01-01')

# İlk 5 satırı göster
print(tesla_data.head())
def make_graph(data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(data['Date'], data['Close'], label='Close Price')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Close Price (USD)')
    plt.legend()
    plt.grid(True)
    plt.show()

# make_graph fonksiyonunu kullanarak grafik çizdirme
tesla_data.reset_index(inplace=True)  # Tarih sütununu tekrar indeks olarak ayarlamak için
make_graph(tesla_data, 'Tesla Stock Price (2020-2023)')
```
